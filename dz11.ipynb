```python
from sympy import Symbol, diff, oo, Interval, is_increasing, is_decreasing
from sympy.solvers import solve
from sympy.plotting import plot
```

```python
x = Symbol('x', real = True)
y = x**2 - 6 * abs(x) + 8
```

```python
print('1. Определить корни:')
roots = solve(y,x)
print(roots)
print('')
```

```python
print('2-3. Найти интервалы, на которых функция возрастает и убывает:')
derivate = diff(y,x)
for i in range(0, len(roots), 2):
    roots_n = []
    roots_n.append(roots[i])
    roots_n.append(roots[i+1])
    roots_n = solve(derivate,x)
    root = roots_n.pop()
    if derivate.subs(x,root + 0.001) > 0: 
        print(f'Function increase in {Interval(root, oo)}')
    else:
        pass
    if derivate.subs(x,root - 0.001) < 0: 
        pass
    else:
        print(f'Function decrease in {Interval(-oo, root)}')
print('')
```
